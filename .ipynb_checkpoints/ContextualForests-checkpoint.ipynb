## Sobre el TFG y distribuciones de probabilidad

Para lo que hablamos de identificar el contexto de una palabra en una frase sin entrenamiento, usamos Wikipedia.
La idea consiste en encontrar todos los posibles significados de una palabra y desarrollar operadores de comparación entre ellos. Wikipedia puede darte todos los articulos con los posibles significados de una palabra; por ejemplo "apple" puede ser "_Apple Inc._","_Apple Records_" "_Apple_"...
Vamos a ir desarrollando el modelo; dada una frase, que tiene $n$ palabras:
$$\{w_i\}_{i \in \{1,...,n\}}$$

Vamos a considerar el conjunto de los artículos de Wikipedia asociados a cada uno de los posibles significados de cada palabra $w_i$: 
$$\mathbb{P}_{i} = \{p^i_1,p^i_2,....,p^i_k\}, \quad p^i_j := \text{Articulo de Wikipedia}$$

El modelo de desambiguación consiste en la creación de un _bosque de desambiguación_, que consiste en un conjunto de árboles que partiendo de una palabra ambigua, se van expandiendo basándose en una heurísitca para buscar conexiones (caminos) entre todas las raices.

La idea es que cada árbol, empieza con una raíz, que es un nodo asociado a una palabra (ambigua o no). Cuando un nodo se expande, sus hijos están formados por nodos asociados a cada uno de los $n$ (parámetro controla la regularidad de los árboles, se modifica en el algoritmo, por defecto son 10) enlaces más relevantes del nodo a expandir.

Cada uno de los nodos tiene un modelo de ajuste de importancia dentro dado por:
$$
    f(w) = P(w \space \text{aparezca en un bloque de 100 palabras tomadas al azar})
$$

Resluta que este modelo sigue una distribución de Zipf y por tanto se puede estimar cual es la importancia $I(\cdot)$ de un enlace compuesto por $k$ palabras, como:
$$
    I(\{w_i\}_{1...k}) = I(\sum_{i=1}^k Ord(w_i))
$$

Donde $ Ord(w_i) $ devuelve la posición de $w_i$ en la lista de palabras ordenadas por importancia en la página a la que pertenece el link.

$$ \max_{(u,v) \in V \times V}||\text{u_links}\space\cap\space\text{v_links}||\sum_{\text{u_links}\space\cap\space\text{v_links}}\min{(I_{u}(l),I_{v}(l))}$$

```python
# Imports and global variables
import wikipediaapi
from wikipedia import search
import spacy
import en_core_web_lg
from nltk.stem.porter import *
from more_itertools import locate
from nltk.stem.snowball import SnowballStemmer
from nltk.stem.porter import PorterStemmer
import string
import numpy as np
from scipy.optimize import curve_fit
from nltk.stem import WordNetLemmatizer 
stemmer = SnowballStemmer(language='english')
lemmatizer = WordNetLemmatizer() 
wiki = wikipediaapi.Wikipedia('en')
nlp = en_core_web_lg.load()
stopwords = "stopwords.txt"
with open(stopwords) as file:
        stop_words = [x.strip() for x in file.readlines()]
```

```python
####################### Interface with Wikipedia ##########################
def _internal_page(title):
    return ":" not in title and "List of" not in title and "(disambiguation)" not in title

def disambiguation(noun):
    """
     Returns all possible pages related with the word
     @Params:
         -noun: name to disambiguate.
     @Returns:
        List with possible options.
    """   
    return [x for x in wiki.page("{} (disambiguation)".format(noun)).links if _internal_page(x)]   

#print(disambiguation("Apple"))

def clean_text(text):
    """Cleans the text, punctuation marks (. , \n) will be replaced by blank spaces 
    and possessive forms ('es , 's) will be removed. 
    Any other punctuation symbol will also be removed.
    @Params:
        -text: to be cleaned
    @Returns:
        Cleaned text
    """
    text = text.replace("\n"," ").replace(","," ").replace("."," ")
    text = text.replace("'s","").replace("'es","")
    text = text.translate(str.maketrans('', '', string.punctuation))
    return text

def stem_text(text):
    #pendiente hacerlo mas elegante
    """
    Uses snowball stemmer to produce a dictionary with entries:
        dic = { ...,key:(words,occurr,relevance),... }
    Where words -> list of ocurrences of key in text (page's text)
          occur -> indexes of (key) occurrences in the text
          relevance -> set to 0 (see set_relevance())
    Note that every key = (STEMMED WORD,WORD_POS (Part Of Speech))
    @Params:
        - text: text to be processed
    @Returns:
        (stem_dictionary,cleaned text length)
    """
    dic = {}
    content = clean_text(text).split() #split spaces
    count = 0
    for i,x in enumerate(nlp(clean_text(text))):
        if x.pos_ not in ["NOUN","PROPN","VERB","ADJ"]:
            continue
        try:
            int(x)
            continue
        except:
            if x.text.lower() not in stop_words and "wikipedia" not in x.text.lower():
                stemmed = stemmer.stem(x.text.lower())
                key = (stemmed,x.pos_)
                if key not in dic:
                    dic[key] = {"words":set([x.text.lower()]),"occurr":[i],"relevance":.0,"pos":0}
                else:
                    dic[key]["words"].add(x.text.lower())
                    dic[key]["occurr"].append(i)
        count += 1
    return dic,len(content)

```

Vamos a considerar como más relevante la palabra que esté más homogéneamente distribuida a lo largo del texto (notar que no tiene por qué ser la palabra más frecuente).

Para ello vamos a considerar una palabra $w$ y un bloque de palabras de tamaño fijo para calcular cuál es la probabilidad de que estando este bloque en cualquier lugar del texto, $w$ esté en él.




```python
def relevance(word,text_len,chunk_size=100):
    """
    Returns word relevance (probability of being in a randomly 
    positioned block of a fixed size).
    @Params:
        - word:  entry of stem_dict
        - text_len: text size (returned by stem_text())
        - chunk_size: block size, fixed
    """
    L = []
    for i in range(chunk_size,text_len,1):
        added = False
        for occ in word["occurr"]:
            if occ > i - chunk_size and occ <= i:
                L.append(1)
                added = True
                break
        if not added:
            L.append(0)
    return np.mean(L)
        
    
def set_relevance(stem_dict,text_len):
    relevances = []
    if text_len < 300: #no aceptamos artículos de menos de 300 palabras
        return None
    else:
        n_chunks = 100 #por ejemplo, mirar otras posibilidades dinamicas etc...
        for word in stem_dict:
            res = relevance(stem_dict[word],text_len,n_chunks)
            stem_dict[word]["relevance"] = res
    return stem_dict
```

```python
import sys

def disorder(l,k):
        sorted_l = sorted(l,key=k)
        diffs = [abs(sorted_l.index(x) - l.index(x)) for x in l]
        return np.mean(diffs)
    
class Node():
    def __init__(self,page,depth,parent=None,root=None,d=20,title=None):
        self.page = page
        self.parent = [parent]
        self.children = {}
        self.d = d
        self.title = title
        if depth == 0:
            self.depth = 0 #raiz, solo para desambiguar
            self.root = None
            self.title = title
        else:
            self.depth = depth
            if self.depth == 1:
                self.root = self
            else:
                self.root = root
            self.dic = set_relevance(*stem_text(self.page.text))
            if self.dic == None: #articulo demasiado corto
                return None
            keys = sorted(self.dic.keys(),key=lambda k:self.dic[k]["relevance"], reverse=True) #palabras de mayor a menot
            y_data = []
            for i,key in enumerate(keys):
                self.dic[key]["pos"] = i
                y_data.append(self.dic[key]["relevance"])
            y_data = np.array(y_data)[y_data != 0][0] 
            x_data = np.linspace(1,len(y_data),len(y_data))
            def zipf(x,alpha): #distribucion de Zipf
                return self.dic[keys[0]]["relevance"]/x**alpha

            popt, pcov = curve_fit(zipf,x_data,y_data)
            self.model = lambda x: zipf(x,popt)
    
    def __clean_links(self,links):
        seen = []
        new_links = []
        for link,page in links:
            ref = link.lower().split(" (")[0].strip()
            if ref not in seen:
                seen.append(ref)
                new_links.append((link,page))
        return new_links
            
    def expand(self):
        if self.depth == 0: #root
            for link,page in self.page.links.items():
                if _internal_page(link) and self.title.lower() in link.lower():
                    self.children[Node(page,self.depth+1,parent=self,root=self.root,title=self.title,d=self.d)] = 1/len(self.page.links)
        else:
            links = sorted(self.page.links.items(),key = lambda l: self.link_relevance(l[0]),reverse=True)
            links = [l for l in self.__clean_links(links) if self.__no_loop(self.page.title,l[0])][:self.d]
            for link,page in links:
                if _internal_page(link):
                    self.children[Node(page,self.depth+1,parent=self,root=self.root,title="",d=self.d)] = self.link_relevance(link)
        return self.children
    
    def __hash__(self):
        return hash(self.page.title)
    
    def __eq__(self, other):
        if other == None:
            return False
        return self.page.title == other.page.title
    
    def __gt__(self, other):
        return self.page.title > other.page.title
    def __no_loop(self,title,link):
        A = title.lower().split()
        B = link.lower().split()
        for a in A:
            for b in B:
                if a in b or b in a:
                    return False
        return True
    
    def link_relevance(self,link):
        words = nlp(link)
        rel = []
        if (self.title.lower() in link.lower() and self.depth == 1) or ":" in link:
                return 0
        for word in words:
            try:
                pos =  self.dic[(stemmer.stem(word.text.lower()),word.pos_)]["pos"]
                rel.append(pos)
            except KeyError:
                rel.append(len(self.dic))
        if len(rel) == 0:
            return 0
        elif sum(rel) == 0:
            return 0
        else:
            m = np.mean(rel)
            return self.model(m)[0]
    
    def similarity(self,other):
        try:
            both = [link for link in self.page.links if link in other.page.links and (other.link_relevance(link) > 0.01 and self.link_relevance(link) > 0.01)]
            #print(both)
            if len(both) == 0:
                return 0
            diff = np.sum([min(self.link_relevance(link),other.link_relevance(link)) for link in both])*len(both)
            return diff
        except:
            return 0
        
        

```

```python
import itertools
from queue import PriorityQueue as pq

class Tree():
    def __init__(self,root_word,exp=10,depth=3):
        word = root_word.lower()
        self.word = word
        #expandes a un primer nivel los posibles significados si los hubiese
        if len(disambiguation(root_word)) != 0:
            self.root = Node(wiki.page("{} (disambiguation)".format(root_word)),0,title = word,d=exp)
            self.root.expand()
            self.expanded,self.to_expand = [self.root],list(self.root.children.keys())
        elif wiki.page(word).exists():
            self.root = Node(wiki.page(word),1,title = word,d=exp)
            self.expanded,self.to_expand = [],[self.root]
        else:
            return
           
    def expand_node(self,node):
        news = node.expand()
        self.expanded.append(node)
        try:
            self.to_expand.remove(node)
        except:
            pass
        self.to_expand += news
        return news
    
    def __hash__(self):
        return hash(self.root.page.title)
    
    def __eq__(self,other):
        return self.root.page.title == other.root.page.title
    
    
class Forest():
    def __init__(self,words):
        self.trees = {}
        self.dic = {}
        for word in words:
            print("Word:{}".format(word))
            self.dic[word] = None
            self.trees[Tree(word)]= None
        self.tree_combs = itertools.combinations(self.trees,2)
        self.connections = {pair:False for pair in self.tree_combs}
        self.Q = pq()
        print("Starting Queue...")
        print("Combinations {}".format(len(list(itertools.combinations(self.trees,2)))))
        for tree1,tree2 in itertools.combinations(self.trees,2):
            #print(tree1.root.page.title,tree1.to_expand,tree2.root.page.title,tree2.to_expand)
            for u in tree1.to_expand:
                for v in tree2.to_expand:
                    sim = -u.similarity(v)
                    if  sim == 0:
                        continue
                    self.Q.put((sim,u,v,tree1,tree2))
    
    def disambiguate(self):
        while not all(self.connections.values()) and not self.Q.empty():
            print("Starting Expansion...")
            _,u,v,t1,t2 = self.Q.get()
            key  =  (t1,t2) if (t1,t2) in self.connections else (t2,t1) #depende de como lo haya metido itertools
            while self.connections[key]:
                _,u,v,t1,t2 = self.Q.get()
                key  =  (t1,t2) if (t1,t2) in self.connections else (t2,t1)
            print("Selected nodes {},{}".format(u.page.title,v.page.title))
            news_t1 = t1.expand_node(u)
            news_t2 = t2.expand_node(v)
           # print(len(set(news_t1).intersection(set(news_t2))))
            #check for intersection
            if any([True if n in t2.to_expand else False for n in t1.to_expand]):
                print("Conexion between trees {},{} established".format(t1.root.title,t2.root.title))
                self.connections[key] = True
                if self.trees[t1] == None:
                    self.trees[t1] = u.root
                if self.trees[t2] == None:
                    self.trees[t2] = v.root
            else: #no connection, add new nodes
                print("Conexion not found, expanding new nodes...")
                #expansion of t1:
                for u in news_t1:
                    for v in news_t2:
                        sim = -u.similarity(v)
                        #print("****",u.page.title,v.page.title,sim)
                        if sim == 0:
                            continue
                        self.Q.put((sim,u,v,t1,t2))
                
                   
                
    def recover_words(self):
        found = False
        for tree,node in self.trees.items():
            for link,page in node.page.links.items():
                if tree.word.lower() == link.lower():
                    self.dic[tree.word] =  Node(page,1)
                    found = True
            if not found:
                connection = node
                while connection.depth != 1:
                    connection = connection.parent
                self.dic[tree.word] = connection
            found = False
            
                    
            
            
                
            
            
        
            
            
        
        
            


                
```

### BERT

En esta parte tomamos 75 frases ambiguas en ingles y comparamos cual es la predicción de BERT para una palabra y cual es la del modelo.

```python
import torch
from transformers import BertTokenizer, BertModel
import matplotlib.pyplot as plt
%matplotlib inline

# Modelo de vocabulario de BERT basico
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

def bert_token(text,word):
    marked_text = "[CLS] " + text + " [SEP]" #marcado final e inicial
    tokenized_text = tokenizer.tokenize(marked_text) #tokens

    indexed_tokens = tokenizer.convert_tokens_to_ids(tokenized_text)
    segments_ids = [1] * len(tokenized_text) #BERT puede analizar varias frases a la vez, los vectores de pertenencia indican indican a que frase pertenecen

    tokens_tensor = torch.tensor([indexed_tokens])
    segments_tensors = torch.tensor([segments_ids]) #sacamos los tokens como tensores de PyTorch

    # modelo preentrenado clasico
    model = BertModel.from_pretrained('bert-base-uncased',output_hidden_states = True)
    # Modo de evaluacion, (se puede poner en entrenamiento para entrenar con tus cosas)
    model.eval()

    with torch.no_grad(): #no construir el grafo de backpropagation

        outputs = model(tokens_tensor, segments_tensors)
        hidden_states = outputs[2] #esto tinene dimension 13(capas de BErt + input)x 1 (frase) x palabras_frase x 768 (tamaño capa de bert)

    token_embeddings = torch.stack(hidden_states, dim=0) #reordenar
    token_embeddings = torch.squeeze(token_embeddings, dim=1)
    token_embeddings = token_embeddings.permute(1,0,2) #reoordenar a palabras x capa x neuronas

    #BERT funciona con representaciones de vectores en las capas ocultas, suele ser una practica
    #común tomar la media de las 4 últimas para los vectores

    token_vecs_sum = []
    for token in token_embeddings:
        sum_vec = torch.sum(token[-4:], dim=0)
        token_vecs_sum.append(sum_vec)

    sentences = {}
    for token,vec in zip(tokenized_text,token_vecs_sum):
        sentences[token.lower()] = vec
    return sentences[word]
```

```python
from scipy.spatial.distance import cosine
import scipy
def get_sentences(word):
    """ Obtiene las primeras frases de los posibles significados"""
    l = {}
    for title,page in wiki.page(word + " (disambiguation)").links.items():
        if ":" not in title and "wiki" not in title and "may refer" not in page.text:
            sep = re.findall('(^.*?[a-z]{2,}[.!?])\s+\W*[A-Z]',page.text)
            if len(sep) == 0:
                continue
            l[title] = sep[0].lower() #primera frase
    return l


def closest(sentence,sentences,word):
    "Calcula dadas las frases de los posibles significados cual es el significado"
    u = bert_token(sentence,word) #obtiene el token de bert de la palabra
    possible,vecs = [],[]
    for meaning,text in sentences.items():
        try:
            vecs.append(bert_token(text,word))
            possible.append(meaning)
        except KeyError: #algo raro hay en el texto o en la tokenizacion de la frase
            continue
    distances = [1-cosine(vec,u) for vec in vecs]
    idx = max(range(len(distances)), key=distances.__getitem__)
    return possible[idx]
    
#sentences = {"Apple Inc.":"Apple technology will take the market",
             #"Apple Records": "Apple is selling a lot of records"}
#u = bert.encode("Apple is selling a lot of phones")
#scipy.spatial.distance.cdist(u,v, "cosine")[0]
```

### GloVe

```python
glove_vecs = {} #vectores de un modelo de BERT entrenado con la wiki, 6B de vectores dim 300
with open("/Users/danielguzman/Downloads/glove/glove.6B.300d.txt") as f:
    lines = f.read().split("\n")[1:]
    for line in lines:
        try:
            word,vec_str = line.split()[0]," ".join(line.split()[1:])
            vec = np.fromstring(vec_str,dtype=float,sep=' ')
            glove_vecs[word] = vec
        except:
            continue
```

```python
def glove_distance(sentence1,sentence2,key):
    tokens1 = [token.text.lower() for token in nlp(sentence1) if token.text.lower() != key.lower()]
    tokens2 = [token.text.lower() for token in nlp(sentence2) if token.text.lower() != key.lower()]
    mx = 0
    i = 0
    for t1 in tokens1:
        for t2 in tokens2:
            try:
                u = glove_vecs[t1]
                v = glove_vecs[t2]
                mx += 1-cosine(u,v)
                i +=1
            except KeyError:
                continue
    return mx/i

def closest(sentence,sentences,word):
    "Calcula dadas las frases de los posibles significados cual es el significado"
    possible,vecs = [],[]
    meanings = list(sentences.keys())
    texts = sentences.values()
    distances = [glove_distance(sentence,text,word) for text in texts]
    idx = max(range(len(distances)), key=distances.__getitem__)
    return meanings[idx]


sentences = get_sentences("apple")
closest("Apple is a computer company",sentences,"apple")
```
